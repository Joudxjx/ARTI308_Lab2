```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

df = pd.read_csv('movies.csv')


if 'Unnamed: 0' in df.columns:
    df = df.drop(columns=['Unnamed: 0'])

df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
df['release_year'] = df['release_date'].dt.year


```

```python
print("\nMissing Values Count:")
print(df.isnull().sum())
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['vote_average'], bins=30, kde=True, color='teal')
plt.title('Distribution of Movie Ratings')
plt.xlabel('Average Rating (0-10)')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
correlation = df[['popularity', 'vote_average', 'vote_count', 'release_year']].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Movie Features')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='popularity', y='vote_average', data=df, alpha=0.4)
plt.title('Movie Popularity vs. Average Rating')
plt.show()
```
